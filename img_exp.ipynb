```python
import cv2
import matplotlib.pyplot as plt

# Read image
ab_img1_bgr = cv2.imread("img1.png", cv2.IMREAD_COLOR)

# Split channels
b, g, r = cv2.split(ab_img1_bgr)

# Display channels
plt.figure(figsize=(20,5))
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title("Red Channel")
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title("Green Channel")
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title("Blue Channel")

# Merge back in RGB order for matplotlib
imgMerged = cv2.merge((r,g,b))
plt.subplot(144); plt.imshow(imgMerged); plt.title("Merged Image")
plt.show()
```

```python
#Chnaging to HSV color space
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("img1.png")
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

h,s,v = cv2.split(img2)

plt.figure(figsize = (20,5))
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H channel")
plt.subplot(142);plt.imshow(s, cmap='gray');plt.title("S channel")
plt.subplot(143);plt.imshow(v, cmap='gray');plt.title("V channel")

img3 = cv2.merge((h,s,v))
plt.subplot(144);plt.imshow(img3);plt.title("Merged Image")
```

```python
#modifying individual channel

import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("img1.png")
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img2)

h_new = h+10
img3 = cv2.merge((h_new,s,v))
img4 = cv2.cvtColor(img3, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(20,5))
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H channel")
plt.subplot(142);plt.imshow(s, cmap='gray');plt.title("S channel")
plt.subplot(143);plt.imshow(v, cmap='gray');plt.title("V channel")

plt.subplot(144);plt.imshow(img4, cmap='gray');plt.title("Merged Image")
```

```python
#saving image

import cv2
from IPython.display import Image
import matplotlib.pyplot as plt

img1 = cv2.imread("img1.png")
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img2)

h_new = h+10
img3 = cv2.merge((h_new,s,v))
img4 = cv2.cvtColor(img3, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(20,5))
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H channel")
plt.subplot(142);plt.imshow(s, cmap='gray');plt.title("S channel")
plt.subplot(143);plt.imshow(v, cmap='gray');plt.title("V channel")

plt.subplot(144);plt.imshow(img4, cmap='gray');plt.title("Merged Image")


#saving process
cv2.imwrite("img2.png", img4)
Image(filename='img2.png')
```
