```python
#drawing a line

import cv2
import matplotlib.pyplot as plt

ab_img = cv2.imread("img1.png")
plt.imshow(ab_img)

gray = cv2.cvtColor(ab_img, cv2.COLOR_BGR2GRAY)

image = gray.copy()

cv2.line(image,(200,200),(400,200),(0,255,255), thickness = 5, lineType=cv2.LINE_AA)

plt.imshow(image, )
```

```python
#drawing a circle
import cv2
import matplotlib.pyplot as plt
ab_img = cv2.imread("img1.png")
plt.imshow(ab_img)

ab_img11 = ab_img.copy()
cv2.circle(ab_img11,(900,500),100,(0,0,255),thickness = 5, lineType=cv2.LINE_AA)

plt.imshow(ab_img11[:,:,::-1])
```

```python
#drawing a rectangle
import cv2
import matplotlib.pyplot as plt
ab_img = cv2.imread("img1.png")
plt.imshow(ab_img)

ab_img11 = ab_img.copy()
cv2.rectangle(ab_img11,(900,500),(100,600),(255,0,255),thickness = 5, lineType=cv2.LINE_8)

plt.imshow(ab_img11[:,:,::-1])
```

```python
#adding a text
import cv2
ab_img = cv2.imread("img1.png")
plt.imshow(ab_img)

ab_image22= ab_img.copy()

text = "Hello this is image"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_COMPLEX
fontColor = (255,255,255)
fontThickness = 5

cv2.putText(ab_image22, text,(200,700), fontFace, fontScale,fontColor,fontThickness, cv2.LINE_AA)

plt.imshow(ab_image22[:,:,::-1])
```
