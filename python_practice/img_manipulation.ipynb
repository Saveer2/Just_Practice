```python
#accessing individual pixels
import cv2
import matplotlib.pyplot as plt

ab_img = cv2.imread("img1.png")
gray = cv2.cvtColor(ab_img, cv2.COLOR_BGR2GRAY)

print(ab_img)
plt.imshow(ab_img)
plt.imshow(gray)

print(ab_img[0,0])
print(ab_img[0,6])
```

```python
#modify image Pixels
import cv2
import matplotlib.pyplot as plt

ab_img = cv2.imread("img1.png")
ab_img_copy = ab_img.copy()
ab_img_copy[2,3] = 200
ab_img_copy[4,5] = 200

plt.imshow(ab_img_copy)
```

```python
#cropping images

import cv2
import matplotlib.pyplot as plt

ab_img = cv2.imread("img1.png")
cropped_region = ab_img[300:500, 600:1000]
plt.imshow(cropped_region)
```

```python
#flipping images

import cv2
import matplotlib.pyplot as plt

ab_img = cv2.imread("img1.png")
ab_img_hori = cv2.flip(ab_img,1)
ab_img_vert = cv2.flip(ab_img,0)
ab_img_both = cv2.flip(ab_img,-1)

plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(ab_img_hori);plt.title("Horizontal flip");
plt.subplot(142);plt.imshow(ab_img_vert);plt.title("Vertical flip");
plt.subplot(143);plt.imshow(ab_img_both);plt.title("Both flip");
```
