```python
#import of lib
import pandas as pd
import matplotlib.pyplot as plt

#load dataset
df = pd.read_csv("titanic_cleaned.csv")
df.columns = df.columns.str.strip().str.lower()

plt.figure(figsize=(10,5))
plt.plot(df['passengerid'], df['fare'], color='blue', linewidth=1)

plt.xlabel("Passenger ID")
plt.ylabel("Fare")
plt.title("Fare Trend Across Passengers")
plt.grid(True)
plt.show()
```
